```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
n = 100000

_mu = 0
_sigma = 1.
x = np.random.normal(_mu,_sigma,(100,n))

x_cs = np.cumsum(x,axis=1)
ns = np.arange(1,n+1) 

mu_estimated = x_cs/ns

sf,fs = 1.5,15
fig,[ax1,ax2] = plt.subplots(1,2,figsize=(sf*2*4,sf*3))

points = np.unique(np.logspace(0,np.log10(n-1),200,base=10,dtype=int))
for i in range(3):
    ax1.plot(ns[0:300],mu_estimated[i,:300],alpha=0.8, label=f"Average-{i+1}")
    ax2.loglog(ns[points],np.abs(mu_estimated[i,points]),alpha=0.8)
    
ax1.set_xlabel("Number of samples in average, n",fontsize=fs)
ax2.set_xlabel("Number of samples in average, n",fontsize=fs)

ax1.set_ylabel("Error, $\mu_{est}-\mu$",fontsize=fs)
ax2.set_ylabel("Absolute error, |$\mu_{est}-\mu$|",fontsize=fs)


ax1.axhline(0,linestyle="--",color="k", label="true  $\mu$")
ax2.loglog(ns,1/np.sqrt(ns),linestyle="--",color="k", label = "$y=n^{-1/2}$")

ax1.legend()
ax2.legend()

ax1.tick_params(labelsize=fs)
ax2.tick_params(labelsize=fs)
plt.tight_layout()
```

```python
from scipy.stats import cauchy

np.random.seed(0)
n = 10000
cauchy_samples = cauchy(loc=0,scale=1).rvs(size=(10,n))

fs,fs = 1.5,15
fig,ax1 = plt.subplots(1,figsize=(sf*4,sf*3))


ns = np.arange(1,n+1)
mu_est = np.cumsum(cauchy_samples,axis=1)/ns
for i in [1,2,6]:
    ax1.plot(ns,mu_est[i,:])
ax1.axhline(0,linestyle="--",color="k",label="loc")

ax1.set_ylim(-3,3)

ax1.set_xlabel("Number of samples in average, n",fontsize=fs)
ax1.set_ylabel("Error, $\mu_{est}-loc$",fontsize=fs)

ax1.legend(fontsize=fs)

ax1.tick_params(labelsize=fs)
```
